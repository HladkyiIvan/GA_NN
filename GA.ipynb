# Simple realisation of genetic algorithm 

### Core DNA class

```python
import random
import string

class DNA:
    
    def __init__(self):
        
        self.genes = ''.join(random.choice(string.ascii_lowercase + ' ') for i in range(len(target)))
        self.fitness = 0
        
    def calculate_fitness(self):
        
        scores = 0
        
        for i in range(len(self.genes)):
            if self.genes[i] == target[i]:
                scores += 1
                
        self.fitness = scores/len(target)
            
    def crossover(self, partner):
        
        child = DNA()
        
        midpoint = random.choice(range(len(self.genes)))
        new_genes = ''
        
        for i in range(len(self.genes)):
            new_genes += self.genes[i] if i < midpoint else partner.genes[i]
        
        child.genes = new_genes
        
        return child
        
    
    def mutate(self):
        
        new_genes = ''
        for i in range(len(self.genes)):
            if random.random() < mutation_rate:
                new_genes += random.choice(string.ascii_lowercase + ' ')
            else:
                new_genes += self.genes[i]
        
        self.genes = new_genes
    
```

### Defining hyperparameters

```python
target = 'i like programming'
mutation_rate = 0.01
total_population = 150

num_of_iterations = 1000
```

### Whole process of algorythm

```python
population = []
mating_pool = []

def setup(target, mutation_rate, total_population):
    
    for i in range(total_population):
        creature = DNA()
        creature.calculate_fitness()
        
        population.append(creature)
        
def draw():
        
    mating_pool = []
        
    for i in range(total_population):
        n = int(population[i].fitness * 100)
        
        for _ in range(n):
            mating_pool.append(population[i])
        
    for i in range(total_population):
        parent_1 = random.choice(range(len(mating_pool)))
        parent_2 = random.choice(range(len(mating_pool)))
        
        child = mating_pool[parent_1].crossover(mating_pool[parent_2])
        
        child.mutate()
        
        population[i] = child
        
    
    for i in range(total_population):
        population[i].calculate_fitness()  
    
        
```

### Start

```python
setup(target, mutation_rate, total_population)

best_result = ''
best_fitness = 0
avg_fitness = 0

for i in range(1, num_of_iterations + 1):
            
    draw()
    
    population.sort(key=lambda c: c.fitness, reverse=True)
    
    
    fitness = [c.fitness for c in population]
    
    if best_fitness < fitness[0]:
        best_fitness = fitness[0]
        best_result = population[0].genes
        
        if best_fitness == 1:
            print('\nWe have a winner!!!')
            print('Best result: ' + best_result)
            print('Best fitness: ' + str(round(best_fitness, 4)))
            break
    
    if i % 20 == 0:
        print('\n--------Best-of-iteration-'+ str(i) +'--------')
        for j in range(1, 6):
            print(str(j) + ': ' + population[j].genes)
        
        print('Average fitness: ' + str(round(sum(fitness)/len(fitness), 4)))
        print('Best result: ' + best_result)
        print('Best fitness: ' + str(round(best_fitness, 4)))
            
        
```
